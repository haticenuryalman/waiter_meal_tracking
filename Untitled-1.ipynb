```python
import os
import face_recognition

garson_klasoru = "garson_fotograflari"
garson_encodingleri = []
garson_isimleri = []


for dosya in os.listdir(garson_klasoru):
    if dosya.endswith(".jpg") or dosya.endswith(".png"):
        yol = os.path.join(garson_klasoru, dosya)
        img = face_recognition.load_image_file(yol)
        enc = face_recognition.face_encodings(img)
        if enc:
            garson_encodingleri.append(enc[0])
            garson_isimleri.append(os.path.splitext(dosya)[0])


def tanimla_garson(goruntu_yolu):
    img = face_recognition.load_image_file(goruntu_yolu)
    enc = face_recognition.face_encodings(img)
    if not enc:
        return "Yüz bulunamadı"
    mesafe = face_recognition.face_distance(garson_encodingleri, enc[0])
    en_yakin = mesafe.argmin()
    return garson_isimleri[en_yakin] if mesafe[en_yakin] < 0.6 else "Eşleşme yok"


print(tanimla_garson("test.png"))
```

```python
import ctypes
ctypes.cdll.LoadLibrary("/opt/homebrew/lib/libzbar.dylib")

import os
os.environ["DYLD_LIBRARY_PATH"] = "/opt/homebrew/lib"

import cv2
import time
import csv
from datetime import datetime
from ultralytics import YOLO
import face_recognition
import numpy as np
import pyzbar.pyzbar as pyzbar
from matplotlib import pyplot as plt
from IPython.display import clear_output

yolo_model = YOLO("food_yolo_best.pt")
garson_faces = {}

def load_garson_faces(klasor="garson_fotograflari"):
    if not os.path.exists(klasor):
        print(f" Klasör bulunamadı: {klasor}")
        return
    for dosya in os.listdir(klasor):
        if dosya.endswith((".jpg", ".png")):
            path = os.path.join(klasor, dosya)
            img = face_recognition.load_image_file(path)
            encs = face_recognition.face_encodings(img)
            if encs:
                garson_faces[os.path.splitext(dosya)[0]] = encs[0]
                print(f" Garson yüklendi: {dosya}")
            else:
                print(f"Yüz bulunamadı: {dosya}")
    print(f"Toplam yüklenen garson sayısı: {len(garson_faces)}")

def recognize_garson(frame, tolerance=0.5):
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    faces = face_recognition.face_locations(rgb)
    encodings = face_recognition.face_encodings(rgb, faces)
    for (top, right, bottom, left), enc in zip(faces, encodings):
        for name, ref_enc in garson_faces.items():
            match = face_recognition.compare_faces([ref_enc], enc, tolerance)[0]
            if match:
                return name, (left, top, right, bottom)
    return "unknown", None

def detect_qr(frame):
    decoded = pyzbar.decode(frame)
    for d in decoded:
        return d.data.decode("utf-8")
    return None

def detect_foods(frame):
    results = yolo_model(frame, conf=0.4)[0]
    items = []
    for r in results.boxes.data.tolist():
        x1, y1, x2, y2, conf, cls = r
        label = yolo_model.names[int(cls)]
        items.append(label)
    return items

def process_video(video_path, save_path=None, show=True, frame_skip_seconds=2):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(" Video açılamadı!")
        return

    load_garson_faces()
    loglar = []

    fps = cap.get(cv2.CAP_PROP_FPS) or 25
    frame_interval = int(fps * frame_skip_seconds)

    out_writer = None
    if save_path:
        fourcc = cv2.VideoWriter_fourcc(*"mp4v")
        w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        out_writer = cv2.VideoWriter(save_path, fourcc, fps, (w, h))

    os.makedirs("yuz_tespit_edilemedi", exist_ok=True)

    frame_id = 0
    aktif_masa = None
    aktif_garson = None
    garson_tanindi = False
    qr_geldi_frame = None

    aktif_yemekler = []
    aktif_hesap = 0

    fiyatlar = {
        "dessert": 300,
        "salad": 200,
        "grilled": 450,
        "soup": 150,
        "stews": 250
    }

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_id += 1

        if frame_id % frame_interval != 0:
            continue

        frame_copy = frame.copy()

        
        yeni_masa = detect_qr(frame)
        if yeni_masa and yeni_masa != aktif_masa:
            
            if aktif_masa and aktif_yemekler:
                now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                loglar.append({
                    "Zaman": now,
                    "Garson": aktif_garson if aktif_garson else "unknown",
                    "Masa": aktif_masa,
                    "Yemekler": aktif_yemekler.copy()
                })
                print(f"Masa değişti! {aktif_masa} → {yeni_masa}, Hesap: ₺{aktif_hesap}, Yemekler: {aktif_yemekler}")
            aktif_yemekler = []
            aktif_hesap = 0
            aktif_masa = yeni_masa
        elif yeni_masa:
            aktif_masa = yeni_masa
        else:
            masa = aktif_masa if aktif_masa else "Bilinmiyor"

      
        garson, bbox = recognize_garson(frame)
        if garson != "unknown":
            aktif_garson = garson
        else:
            garson = aktif_garson if aktif_garson else "unknown"

       
        if bbox:
            l, t, r, b = bbox
            cv2.rectangle(frame, (l, t), (r, b), (0, 255, 0), 2)
            cv2.putText(frame, garson, (l, t - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
        else:
            if garson == "unknown":
                cv2.imwrite(f"yuz_tespit_edilemedi/frame_{frame_id:04}.jpg", frame)

        
        if qr_geldi_frame and not garson_tanindi:
            if (frame_id - qr_geldi_frame) >= 30:
                print(f"[UYARI] {aktif_masa} masasına 60 saniyedir garson gelmedi!")
                garson_tanindi = "uyarı_verildi"

      
        yemekler = detect_foods(frame_copy)
        for yemek in yemekler:
            if not aktif_yemekler or aktif_yemekler[-1] != yemek:
                aktif_yemekler.append(yemek)

       
        aktif_hesap = sum(fiyatlar.get(y, 0) for y in aktif_yemekler)



       
        results = yolo_model(frame_copy, conf=0.4)[0]
        for r in results.boxes.data.tolist():
            x1, y1, x2, y2, conf, cls = r
            label = yolo_model.names[int(cls)]
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 0, 255), 2)
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

        
        cv2.putText(frame, f"Masa: {aktif_masa if aktif_masa else 'Bilinmiyor'}", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 255), 2)
        cv2.putText(frame, f"Garson: {garson}", (20, 80), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
        cv2.putText(frame, f"Yemekler: {', '.join(aktif_yemekler)}", (20, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        cv2.putText(frame, f"Hesap: ₺{aktif_hesap}", (20, 160), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

        
        if show:
            clear_output(wait=True)
            plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            plt.title(f"Frame: {frame_id}")
            plt.axis("off")
            plt.show()
            time.sleep(0.2)

        if out_writer:
            out_writer.write(frame)

    
    if aktif_masa and aktif_yemekler:
        now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        loglar.append({
            "Zaman": now,
            "Garson": aktif_garson if aktif_garson else "unknown",
            "Masa": aktif_masa,
            "Yemekler": aktif_yemekler.copy()
        })

    cap.release()
    if out_writer:
        out_writer.release()

    with open("loglar.csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["Zaman", "Garson", "Masa", "Yemekler"])
        writer.writeheader()
        for row in loglar:
            row["Yemekler"] = ", ".join(row["Yemekler"])
            writer.writerow(row)

    print(" İşlem tamamlandı. Log kaydedildi → loglar.csv")


process_video("deneme.mov", save_path="cikti_video.mp4", show=True, frame_skip_seconds=2)
```
